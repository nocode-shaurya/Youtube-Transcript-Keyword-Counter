```python
from yt_transcript_fetcher import list_languages, get_transcript

video_id = "5UxmKmlxiuY"  # Replace with your target video ID

# List all available languages
languages = list_languages(video_id)
print("Available languages:", languages)

transcript_segments = get_transcript(video_id, "hi")
transcript = []
if hasattr(transcript_segments, 'segments'):
    for segment in transcript_segments.segments:
        transcript.append(segment.text)
else:
    print("No segments found in the transcript object.")

print(transcript)
```

```python
import re
search_word = input("Give text in Hindi to count occurrences of: ")
pattern = re.escape(search_word)

modi_count = 0
for i in transcript:
    modi_count += len(re.findall(pattern, i))


print(f"Occurrences of '{pattern}': {modi_count}")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
import re

keyword = "Today Breaking News"


driver = webdriver.Chrome()
driver.get('https://www.youtube.com/c/@DLSNews/videos')
time.sleep(5)  # Wait for page to load (increase if needed)
scroll_pause = 2
for _ in range(1):  # Adjust for how many videos you want
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(scroll_pause)

video_links = driver.find_elements(By.XPATH, '//a[@id="video-title-link"]')

video_ids = []
for link in video_links:
    title = link.get_attribute('title')
    href = link.get_attribute('href')
    # print('Video Title:', title)
    # print('Video Link:', href)
    if re.search(keyword, title, re.IGNORECASE) != None:
        print('Keyword found in title:', title)
        if href and 'v=' in href:
            video_id = href.split('v=')[1]
            video_ids.append(video_id)
    else:
        continue


print()
print(video_ids)
print(len(video_ids))
driver.quit()
```
